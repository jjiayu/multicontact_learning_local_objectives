## Script for Collecting Data Points

```python
#Import Packages
import numpy as np
import os
import pickle
from multicontact_learning_local_objectives.python.machine_learning.ml_utils import *
import matplotlib.pyplot as plt #Matplotlib
import shutil

print("Double check the Path for storing trajectories is correct")
```

```python
#Define Path for Storing Trajectories
#Collect Data Points Path
#workingDirectory = "/home/jiayu/Desktop/multicontact_learning_local_objectives/data/large_slope_flat_patches/"
workingDirectory = "/home/jiayu/Desktop/MLP_DataSet/antfarm_2steps"
#"/home/jiayu/Desktop/multicontact_learning_local_objectives/data/large_slope_flat_patches/"
#NOTE: need to have "/" at the end
print("Double Check we provide the Correct Traj Path: \n", workingDirectory)

#Define Roll Out Path
rolloutPath = workingDirectory + "/CleanTrainingSetRollOuts/"

#create dataset folder if we dont have one
DataSetPath = workingDirectory + "/DataSet/"
if not (os.path.isdir(DataSetPath)):
    os.mkdir(DataSetPath)

#create the Machine learing folder
#Path to store ML Model
ML_Model_Path = workingDirectory + "/ML_Models/"
if not (os.path.isdir(ML_Model_Path)):
    os.mkdir(ML_Model_Path)
```

```python
#Shift world frame type: StanceFoot, (Abandon) InitCoM,  (Abandon) InitSurfBorder, None
Shift_World_Frame_Type = "StanceFoot";       print("Shift World Frame to:\n", Shift_World_Frame_Type)
#Contact Location Representation Type 1) 3DPoints 2) ConvexCombination 3) FollowRectangelBorder
Contact_Representation_Type = "FollowRectangelBorder";   print("Contact Location Representation Type:\n ", Contact_Representation_Type)
#Scaling Vectors? 1 menas no scaling (Now scaling does not make much diff)
ScaleFactor = 1;      print("Scaling Factor of all quantities (Except Left and Right Swing Flag):\n", ScaleFactor)
```

```python
#Start Collecting Data Points

#get all the file names
filenames = os.listdir(rolloutPath)

#Initialize Dataset
x_all = [];    y_all = []

fig, ax = plt.subplots()

#Count Success full 
total_file_num = 0
success_file_num = 0

#loop all the files
for filename in filenames:
    if ".p" in filename:#a data file
        print("Process: ",filename)

        total_file_num = total_file_num + 1 #get a file

        #Load data
        with open(rolloutPath+filename, 'rb') as f:
            data= pickle.load(f)

        if len(data["SingleOptResultSavings"]) == data["Num_of_Rounds"]:
            
            #This is successful Computation
            print("Success Computation")
            success_file_num = success_file_num + 1
            
            #print Terrain Settings
            print(data["TerrainSettings"])

            #for roundNum in range(1):
            for roundNum in range(data["Num_of_Rounds"]):
                print("   Process Round: ", roundNum)
                #Get Single Optimization Result of current result
                singleOptResult = data["SingleOptResultSavings"][roundNum]
                
                #Get data Points
                #xtemp, ytemp = getDataPoints(SingleOptRes = singleOptResult, Shift_World_Frame = None, ContactRepresentationType = Contact_Representation_Type, VectorScaleFactor = ScaleFactor) #InitCoM; InitSurfBorder
                #print("None Shifted Frame")
                #print(xtemp);   print(ytemp)

                xtemp, ytemp = getDataPoints(SingleOptRes = singleOptResult, Shift_World_Frame = Shift_World_Frame_Type, ContactRepresentationType = Contact_Representation_Type, VectorScaleFactor = ScaleFactor) #InitCoM; InitSurfBorder
                #print(xtemp.shape)
                #print("with Shifted Frame")
                #print(xtemp);   print(ytemp)
                x_all.append(xtemp);    y_all.append(ytemp)

                ax.scatter(xtemp[0], xtemp[4], marker='o', linewidth = 5) 
        else:
            print("Failed Computation")  

#make data points to become numpy array
x_all = np.array(x_all);    y_all = np.array(y_all)
#print(x_all.shape)

plt.show()


#Get NumberofPreviewSteps
NumPreviewSteps = len(data["SingleOptResultSavings"][0]["ContactSurfs"])

#Save DataPoints and Settings
DatasSet = {"input":  x_all,
            "output": y_all,
            "Shift_World_Frame_Type":Shift_World_Frame_Type,
            "Contact_Representation_Type": Contact_Representation_Type,
            "VectorScaleFactor":ScaleFactor,
            "NumPreviewSteps": NumPreviewSteps}

print("DataSet Settings:")
print("- Shift_World_Frame_Type: ", DatasSet["Shift_World_Frame_Type"])
print("- Contact Location Representation Type: ", DatasSet["Contact_Representation_Type"])
print("- Vector Scale Factor: ", DatasSet["VectorScaleFactor"])
print("- Number of Preview Steps: ", DatasSet["NumPreviewSteps"])

#Save the data
pickle.dump(DatasSet, open(DataSetPath + "/data"+'.p', "wb"))

print(" ")
print("Summary of Data Point Collection: ")
print("- Total Number of Rollouts (files): ", total_file_num)
print("- Total Number of Success Rollouts (files): ", success_file_num)
```

```python

#Learning Code
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Load DataSet File

dataset_file = DataSetPath + "/data"+'.p'
dataset = pickle.load(open(dataset_file,"rb"))

print("World Frame Shift: ", dataset["Shift_World_Frame_Type"])
print("Contact Location Representation Type: ",dataset["Contact_Representation_Type"])
print("Scaling Factor of Variables: ",dataset["VectorScaleFactor"])
print("Number of Preview Steps: ", dataset["NumPreviewSteps"])
```

```python
#Test Train Split, for now No Test data
#x_train, x_test, y_train, y_test = train_test_split(dataset["input"], dataset["output"], test_size = 0.01)
x_train = dataset["input"]
y_train = dataset["output"]

#Decide input and outpu dimensionality
d_in = x_train[0].shape[0]
d_out = y_train[0].shape[0]
```

```python
#Define learning model
model = Sequential([
    Dense(256, activation='relu', input_shape=(d_in,)),
    Dense(256, activation='relu', input_shape=(d_in,)),
    Dense(256, activation='relu', input_shape=(d_in,)),
    Dense(256, activation='relu', input_shape=(d_in,)),
    Dense(d_out)
])
```

```python
#Train Learning Model
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001), loss='mse')

history = model.fit(x_train, y_train, epochs = 1000, validation_split=0.0, batch_size = x_train.shape[0])
#history = model.fit(x_train, y_train, epochs = 100000, validation_split=0.2)
```

```python
#Plot Training Progree
plt.plot(history.history['loss'], 'r', label='loss')
plt.yscale("log")
plt.plot(history.history['val_loss'], 'b', label='val_loss') if 'val_loss' in history.history else None
plt.legend()
plt.show()
```

```python
#Save Trained Model
MLmodel_name = "NN_Model_Valid"
model.save(ML_Model_Path + MLmodel_name)
```

```python
#Save DataSet Setttings
datasetSettings = {"Shift_World_Frame_Type":dataset["Shift_World_Frame_Type"],
                   "VectorScaleFactor":dataset["VectorScaleFactor"],
                   "NumPreviewSteps":dataset["NumPreviewSteps"],
                   "Contact_Representation_Type":dataset["Contact_Representation_Type"],
                   "TrainingLoss":history.history['loss']}
datasetSettings["ValidationLoss"] = history.history['val_loss'] if 'val_loss' in history.history else None
pickle.dump(datasetSettings, open(ML_Model_Path + MLmodel_name+ '/datasetSettings' +'.p', "wb"))
```

```python
#Show Prediction Result for Training
datapoint_num = 40
y_pred_train = model.predict(np.array([x_train[datapoint_num]]))

print(x_train[datapoint_num])

print("predicted result: \n",y_pred_train/dataset["VectorScaleFactor"])
print("true value: \n",np.array([y_train[datapoint_num]])/dataset["VectorScaleFactor"])
print("diff: \n", (y_pred_train - np.array([y_train[datapoint_num]]))/dataset["VectorScaleFactor"])
```
